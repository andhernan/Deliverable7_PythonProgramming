Create a Python program that compiles the sequence of the first non-negative n even, odd, and Fibonacci numbers.

```python
#    Ask the user to enter an integer value greater than 0.
n = int(input("Enter an non-negative integer number: "))
```

```python
# Use a conditional statement to check the userâ€™s entry is greater than 0

# I was going to go crazy with a while statement and try's 
# BUT THEN I realized a simple while will do
# #almostFumbled
while n <= 0:
  print("Amount is not greater than zero. \n Try Again")
  n = int(input("Enter an non-negative integer number: "))
```

```python
# Create a list for each sequence, and store the numbers of sequences. 
# Lists should have the same length,  n (user's entry).
even = []
odd = []
fibo = [0,1]

for i in range(n):
  even.append(2*i)
  odd.append(2*i+1)
  if i > 1: fibo.append(fibo[i-2]+fibo[i-1])
```

```python
# Display your results in a data frame format.

fib_dict = {'Even':even, 'Odd': odd, 'Fibonacci': fibo}

fib_df = pd.DataFrame(data = fib_dict)

fib_df
```
